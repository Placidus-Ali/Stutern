```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
# read anscombe's dataset
df = pd.read_csv('anscombe_quartet_data.csv')
df.head()
```

```python
# Defining my axes
x1 = df.x1
y1 = df.y1
x2 = df.x2
y2 = df.y2
x3 = df.x3
y3 = df.y3
x4 = df.x4
y4 = df.y4
```

```python
# Ploting the my Subplots using the 1st method

#Defining a function to add diagonal line in my subplots
def add_diagonal_line(x, y): 
    x, y = [min(x), max(x)], [min(y), max(y)]
    plt.plot(x, y)

plt.subplot(2, 2, 1)
plt.title('Graph of x1, y1')
plt.scatter(x1, y1, color='green')
plt.xlabel('x1')
plt.ylabel('y1')
add_diagonal_line(x1,y1)

plt.subplot(2, 2, 2)
plt.title('Graph of x2, y2')
plt.scatter(x2, y2, color='red')
plt.xlabel('x2')
plt.ylabel('y2')
add_diagonal_line(x2,y2)

plt.subplot(2, 2, 3)
plt.title('Graph of x3, y3')
plt.scatter(x3, y3, color='yellow')
plt.xlabel('x3')
plt.ylabel('y3')
add_diagonal_line(x3,y3)

plt.subplot(2, 2, 4)
plt.title('Graph of x4, y4')
plt.scatter(x4, y4, color='black')
plt.xlabel('x4')
plt.ylabel('y4')
add_diagonal_line(x4,y4)

plt.subplots_adjust(wspace=0.7, hspace=0.7, bottom=0.1, top=0.9)
plt.show()
```

```python
# Ploting the my Subplots using the 2nd method

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

ax1.scatter(x1, y1, color='green')
ax1.set_xlabel('x1')
ax1.set_ylabel('y1')
ax1.plot(ax1.get_xlim(), ax1.get_ylim())
ax1.set_title('Graph of x1, y1')
ax2.scatter(x2, y2, color='red')
ax2.set_xlabel('x2')
ax2.set_ylabel('y2')
ax2.plot(ax2.get_xlim(), ax2.get_ylim())
ax2.set_title('Graph of x2, y2')
ax3.scatter(x3, y3, color='yellow')
ax3.set_xlabel('x3')
ax3.set_ylabel('y3')
ax3.plot(ax3.get_xlim(), ax3.get_ylim())
ax3.set_title('Graph of x3, y3')
ax4.scatter(x4, y4, color='black')
ax4.set_xlabel('x4')
ax4.set_ylabel('y4')
ax4.plot(ax4.get_xlim(), ax4.get_ylim())
ax4.set_title('Graph of x4, y4')


plt.subplots_adjust(wspace=0.7, hspace=0.7, bottom=0.1, top=0.9)
```

```python
# Ploting the my Subplots using the 3rd method

fig = plt.figure(figsize=(8, 6))

ax1 = fig.add_subplot(221, title='Graph of x1, y1')
ax1.set_xlabel('x1')
ax1.set_ylabel('y1')
ax1.scatter(x1, y1, color='green')
ax1.plot(ax1.get_xlim(), ax1.get_ylim())
ax2 = fig.add_subplot(222, title='Graph of x2, y2')
ax2.set_xlabel('x2')
ax2.set_ylabel('y2')
ax2.scatter(x2, y2, color='red')
ax2.plot(ax2.get_xlim(), ax2.get_ylim())
ax3 = fig.add_subplot(223, title='Graph of x3, y3')
ax3.set_xlabel('x3')
ax3.set_ylabel('y3')
ax3.scatter(x3, y3, color='yellow')
ax3.plot(ax3.get_xlim(), ax3.get_ylim())
ax4 = fig.add_subplot(224, title='Graph of x4, y4')
ax4.set_xlabel('x4')
ax4.set_ylabel('y4')
ax4.scatter(x4, y4, color='black')
ax4.plot(ax4.get_xlim(), ax4.get_ylim())


plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```

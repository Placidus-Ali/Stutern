```python
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
```

```python
# Read the csv file
df = pd.read_csv('company_sales_data.csv')
df
```

```python


plt.plot(df.month_number,df.facecream, color= 'olive', linestyle= '--', marker= "o")
plt.plot(df.month_number,df.facewash, color= 'black', linestyle= '--', marker= "d")
plt.plot(df.month_number,df.toothpaste, color= 'blue', linestyle= ':', marker= "o")
plt.plot(df.month_number,df.bathingsoap, color= 'purple', linestyle= '-', marker= "X")
plt.plot(df.month_number,df.shampoo, color= 'indigo', linestyle= ':', marker= "o")
plt.plot(df.month_number,df.moisturizer, color= 'red', linestyle= '', marker= "o")
plt.title("A Multiple Line Graph Of Months Against Product Sales")
plt.xlabel("Months Of The Year In Numbers")
plt.ylabel("Amount Of Sales From Products")
plt.legend(["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"], loc='upper left')
plt.show()
plt.savefig('company_sales_data.png')
```

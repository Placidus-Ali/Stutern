```python
import numpy as np
import pandas as pd
```

After importing all the necessary libraries, I read the 'movies.csv' file showing just the head.

```python
# read movies csv file
df = pd.read_csv('movies.csv')
df.head()
```

Here I tried to check and confirm all the datatypes for each column

```python
# checking the datatype of all the columns
df.dtypes
```

Here I checked for missing values. Then I found that there are some missing values in columns like YEAR, GENRE, RATING, VOTES, RunTime Aand Gross. 

```python
# checking for missing values
df.isnull().any()
```

Here I replaced the missing values with str '2021'

```python
# replacing null values in YEAR Column
df['YEAR'].fillna('2021', inplace=True)
df.head()
```

At the GENRE column I noticed that the most common GENRE is '\nAnimation, Action, Adventure' and it's str, so I replaced the null values there with it.

```python
# filling the GENRE column
df['GENRE'].fillna('\nAnimation, Action, Adventure', inplace=True)
df.head()
```

RATINg has so many missong values, but then I tried replacing them with the mean of the available values since its datatype is float64.

```python
# filling the RATING column

x = df['RATING'].mean()
df['RATING'].fillna(x, inplace=True)
df.tail()
```

The VOTES column has many missing values as well but then it's a str datatype, so using mean without converting the datatype will not work so I checked for the most commoon value in the column ('414,849') and used it to replace the missing ones.

```python
# filling the VOTES column

df['VOTES'].fillna('414,849', inplace=True)
df.head()
```

The RunTime datatype is float64 so it was abit easier to check and use the mean of the values to fill the missing values.

```python
# filling NaN values in RunTime column
x = df['RunTime'].mean()
df['RunTime'].fillna(x, inplace=True)
df.head()
```

Gross column has the highest number of missing values, but then it's a str datatype. I tried converting the datatype to float or int so I can use the mean or mode of the values to fill the missing ones, but then it didn't work for me. So decided to use a avarage number '150.70M' to fill all the missing values.

```python
# filling NaN values in Gross column
df['Gross'].fillna('150.70M', inplace=True)
df.head()
```

Here, after filling all the missing values in all the columns, I tried to reconfirm and they all showed false.

```python
# checking if I still have null value in the whole DataFrame. Good, I don't have!
df.isnull().any()
```

The next thing I did was check for deplicates and it shpwed True.

```python
# checking for duplicate row
df.duplicated().any()
```

Then I tried to confirm how many rows where duplicated and I got about 431 rows

```python
# checking for how duplicate row
df.duplicated().sum()
```

Here I just tried to cofirm if all the columns has duplicates so i will know the how to apply the drop_duplicates function.

```python
# showing the duplicates row
df[df.duplicated(['MOVIES'], keep=False)].sort_values(by=['MOVIES']).sum()
```

Here I dropped all the 431 duplicates, keeping the first row before the duplicates and ignoring the index so the table will still be well arranged.

```python
# dropping duplicate rows
df.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=False)
df.head()
```

```python
# confirming if there are still dupliacte rows
df.duplicated().any()
```

```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Read company sales data
df = pd.read_csv('company_sales_data.csv')
df
```

```python
# Defining my variables
toothpaste = df.toothpaste
moisturizer = df.moisturizer
x = df.month_number
bathingsoap = df.bathingsoap
```

A bar plot comparing the sales between two products over the year.
This graphs shows that Toothpaste sold far more than Moisturizer in all months of the year. Especially on the 10th month where Toothpaste sold over $8,000 and Moisturizer sold just $1,890. From the graph, Moisturizer's highest sale was $2,100 in the 11th month, and that is not even upto half of the least Toothpaste sales ($4,550) in the 3rd month of the year. 

```python
#ploting a side by side bar chart to compare the 

# setting the width of my bars 
fig = plt.subplots(figsize =(12, 8))

month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']


plt.bar(x +0.2, toothpaste, color ='red', width = 0.40)
plt.bar(x -0.2, moisturizer, color ='green', width = 0.40)
plt.xlabel('Months Of The Year')
plt.ylabel('Products Sales for Toothpaste and Moisturizer')
plt.legend(['toothpaste','moisturizer'], loc= 'upper left')
plt.title('Comparing The Sales Between Toothpaste And Moisturizer Over 1Year')
```

```python

error = np.array(list(map(lambda x: 10000-x, df.bathingsoap)))
yerror = np.absolute(error)
print(yerror)
plt.bar(x, bathingsoap, yerr=yerror, color='red', capsize=10)
plt.title('Bathingsoap Monthly Sales')
plt.ylabel('Sales')
plt.xlabel('Months of the Year')
plt.legend(['bathingsoap'], loc='upper left', fontsize=14)
plt.show()
```

From the pie chart, we can see that Bathingsoap sold highest over the year with $114,010 total sales, followed by Toothpaste with $69,910 total sales. Facewash and Moisturizer had the least total sales with $18,515 each.

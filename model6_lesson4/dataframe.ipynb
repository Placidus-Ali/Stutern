Q5: Perform an outer merge between customers and products

Here the result displayed all the rows in both DataFrames but then returned NaN for those rows that doesn't have values in the Products Table.

```python
Customers_Product = pd.merge(Customers, Products, left_index=True, right_index=True, how="outer")
Customers_Product
```

Q6: Perform a left merge between orders and customers

Here it merged the customer_id column in the two DataFrames. But then, it first showed the Orders Table at the left and the Customers Tabel at the right. Again it returned NaN for the remaining rows in the Customers DataFrame that doesn't values since the Orders Dataframes has 20 rows while Customers DataFrame has just 10

```python
Orders_Customers = pd.merge(Orders, Customers, how="left")
Orders_Customers
```

Q7: Perform a right merge between product and customers

```python
Products_Customers = pd.merge(Products, Customers, left_index=True, right_index=True,  how="right")
Products_Customers
```

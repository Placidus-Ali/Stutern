Importing all the necessary libraries

```python
import matplotlib 
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

 Loading and chceking my dataframe called df

```python
df = pd.read_csv('WorldCupMatches.csv')
df.head()
```

Checking the data type of the columns in the Dataframe 

```python
df.dtypes
```

Checking if there are missing values in the columns in the DataFrame

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

Droping missing values in my DataFrame before visualizing to avoid error

```python
df.dropna(how='any', inplace=True)
df
```

Confirming that there are no more missing values in my DataFrame

```python
df.isnull().any()
```

Creating new column called Total Goals to house the sum of Home Team Goals and Away Team Goals

```python
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.head()
```

Visualizing the Total Goals scored in each match using the displot type

```python
sns.displot(df['Total Goals'], color='red').set(title= 'A Plot of Total Number of Goals in each Match')
sns.set_style("whitegrid")
plt.show()
```

Q5&6

```python
fig, ax = plt.subplots(figsize=(12, 7))
sns.set_style('whitegrid')
sns.barplot(x= df.Year, y= df['Total Goals'], ci=None, estimator=sum)
plt.title('Total Number of Goals Scored Per Year')
plt.show()
```

Visualizing the columns Year and Total Goals as a bar chart. Putting Year on the x-axis, and Total Goals on the y-axis.

```python
sns.barplot(x= df.Year, y= df['Total Goals'], ci=None)
plt.title('World Cup Match Goals')
plt.xticks(rotation='vertical')
plt.show()
```

Q9&10

```python
fig, ax = plt.subplots(figsize=(12, 7))
sns.boxplot(x= df.Year, y= df['Total Goals'])
plt.title('Average Yearly Goals')
plt.show()
```
